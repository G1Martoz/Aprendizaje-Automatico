# Actividad: Importación y Análisis de Datos con Pandas

Esta actividad consiste en importar datos desde diferentes formatos (CSV, JSON, Excel) y realizar operaciones básicas de análisis.

```python
import pandas as pd
import numpy as np
import os

# No necesitamos BASE_DIR en Jupyter si los archivos están en la misma carpeta,
# pero lo mantenemos por compatibilidad si se ejecuta desde otro directorio.
try:
    # En Jupyter, la ruta base es el directorio del notebook
    BASE_DIR = os.getcwd()
except:
    BASE_DIR = os.path.dirname(os.path.abspath(__file__))
```

### 1. Carga de datos de Ventas (CSV)
Cargamos el archivo `ventas.csv` y mostramos las primeras filas.

```python
try:
    path_ventas = os.path.join(BASE_DIR, 'ventas.csv')
    ventas_df = pd.read_csv(path_ventas, encoding="ISO-8859-1")
    print("--- Punto 1: Vista previa de Ventas ---")
    display(ventas_df.head())
except FileNotFoundError:
    print(f"Error: El archivo ventas.csv no se encuentra en {BASE_DIR}")
```

### 2. Carga de datos de Clientes (JSON)
Cargamos el archivo `clientes.json` y mostramos la cantidad total de clientes.

```python
try:
    path_clientes = os.path.join(BASE_DIR, 'clientes.json')
    clientes_df = pd.read_json(path_clientes)
    print("--- Punto 2: Información de Clientes ---")
    print(f"Cantidad total de clientes registrados: {len(clientes_df)}")
except FileNotFoundError:
    print(f"Error: El archivo clientes.json no se encuentra en {BASE_DIR}")
```

### 3. Carga de Inventario (Excel)
Cargamos el archivo `inventario.xlsx` y calculamos el promedio de stock.

```python
try:
    path_inventario = os.path.join(BASE_DIR, 'inventario.xlsx')
    inventario_df = pd.read_excel(path_inventario) 
    
    print("--- Punto 3: Análisis de Inventario ---")
    promedio_stock = inventario_df['stock'].mean()
    print(f"El promedio de stock disponible es: {promedio_stock:.2f}")
except FileNotFoundError:
    print(f"Error: El archivo inventario.xlsx no se encuentra en {BASE_DIR}")
except KeyError:
    print("Error: No se encontró la columna 'stock' en el archivo de inventario.")
```
