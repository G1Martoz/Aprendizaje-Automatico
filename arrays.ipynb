```python
import numpy as np
np.random.seed(47)
```

```python
# Ejercicio 1: Array de 10 elementos con cero
arr1 = np.zeros(10)
print(arr1)
```

```python
# Ejercicio 2: Array con numeros del 1 al 20
arr2 = np.arange(1, 21)
print(arr2)
```

```python
# Ejercicio 3: Array con numeros pares del 2 al 50
arr3 = np.arange(2, 51, 2)
print(arr3)
```

```python
# Ejercicio 4: Array con numeros impares del 1 al 50 (supongo que quiso decir impares)
arr4 = np.arange(1, 51, 2)
print(arr4)
```

```python
# Ejercio 5: Matriz 4x4 con valores aleatorios entre 0 y 1:
arr5 = np.random.rand(4, 4)
print(arr5)
```

```python
# Ejercicio 6: Suma elemento a elemento:
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([[7, 8, 9], [10, 11, 12]])
arr1 + arr2
```

```python
# Ejercicio 7: Multiplica dos matrices 3x3 de enteros aleatorios entre 1 y 10
# -Ejemplo de algebra lineal .T intercambia filas por columnas
m1 = np.random.randint(1, 11, (3, 3))
m2 = np.random.randint(1, 11, (3, 3))
print(m1)
print(m2)
print(m1 @ m2.T)
```

```python
# Ejercicio 7 con multiplicacion de matrices 3x3 de enteros aleatorios entre 1 y 10
m1 = np.random.randint(1, 11, (3, 3))
m2 = np.random.randint(1, 11, (3, 3))
mr1 = np.dot(m1, m2)
print(f"El producto de las matrices 3zx3 es:\n{mr1}")
```

```python
# Ejercicio 8: Array de 15 valores aleatorios (1-100) e imprime estadisticos:
arr8 = np.random.randint(1, 101, 15)
print(arr8)
print(f"Mximo : {np.max(arr8)} Minimo : {np.min(arr8)} Promedio : {np.mean(arr8)}")
```

```python
# Ejercicio 9: Genera una matriz 5x5 con valores del 1 al 25 y extraer su diagonal.
m5x5 = np.arange(1, 26).reshape(5, 5)
diagonal = np.diag(m5x5)
print(m5x5)
print(f"La diagonal de la matriz 5x5 es:\n{diagonal}")
```

```python
# Ejercicio 10: Matrix 5x5 (1-20), y calcular suma de cada fila y columna.
mf1 = np.random.randint(1, 21, (5, 5))
suma_filas = mf1.sum(axis=1)
suma_columnas = mf1.sum(axis=0)
print(f"Matriz aleatoria:\n{mf1}")
print(f"Suma de cada fila:\n{suma_filas}")
print(f"Suma de cada columna:\n{suma_columnas}")
```
